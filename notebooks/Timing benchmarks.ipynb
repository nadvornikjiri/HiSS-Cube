```python
import pandas as pd
import plotly.express as px
import datetime

df = pd.read_csv("../results/benchmark/default_baseline.csv")
df.rename({"Time": "Baseline"}, axis=1, inplace=True)
df2 = pd.read_csv("../results/benchmark/serialized_fits_headers.csv")
df2.rename({"Time": "Serialized Headers"}, axis=1, inplace=True)
df3 = pd.read_csv("../results/benchmark/serialized_fits_headers_2.csv")
df3.rename({"Time": "Serialized Headers 2"}, axis=1, inplace=True)
df4 = pd.read_csv("../results/benchmark/serialized_fits_headers_3.csv")
df4.rename({"Time": "Serialized Headers 3"}, axis=1, inplace=True)

df_stacked = pd.concat([df, df2["Serialized Headers"]], axis=1)
df_stacked = pd.concat([df_stacked, df3["Serialized Headers 2"]], axis=1)
df_stacked = pd.concat([df_stacked, df4["Serialized Headers 3"]], axis=1)

fig = px.scatter(df_stacked, x="Image count", y=df_stacked.columns[2:6], title="Processing time", trendline="rolling", 
                 trendline_options=dict(window=5))
fig.show()

fig2 = px.line(df_stacked, x="Image count", y=df_stacked.columns[2:4], title="Processing time")
fig2.show()

df_subset = df_stacked.loc[df_stacked["Image count"] > 30000]

fig3 = px.scatter(df_subset, x="Image count", y=df_subset.columns[2:4], title="Processing time", trendline="rolling", 
                 trendline_options=dict(window=5))
fig3.show()

# Total time
print("Baseline time: %s" % str(datetime.timedelta(seconds=df_stacked["Baseline"].sum())))
print("Serialized Headers time: %s" % str(datetime.timedelta(seconds=df_stacked["Serialized Headers"].sum())))
print("Serialized Headers 2 time: %s" % str(datetime.timedelta(seconds=df_stacked["Serialized Headers 2"].sum())))


```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("../results/benchmark/default_baseline.csv")

fig = px.scatter(df, x="Image count", y="Time", title="Processing time", trendline="rolling", 
                 trendline_options=dict(window=5))
fig.show()

fig2 = px.line(df, x="Image count", y="Time", title="Processing time")
fig2.show()

df_subset = df.loc[df["Image count"] > 30000]

fig3 = px.scatter(df_subset, x="Image count", y="Time", title="Processing time", trendline="rolling", 
                 trendline_options=dict(window=5))
fig3.show()

df["Time"].sum() / 3600
```

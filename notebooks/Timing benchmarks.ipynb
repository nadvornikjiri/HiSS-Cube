```python
import pandas as pd
import plotly.express as px
import datetime
import os 
import glob

extension = 'csv'
logs_path = '../results/benchmark/timing_logs/without_cache_drop'
timing_types=["groups_only", "groups_datasets", "groups_attrs_datasets"]
no_iterations = 3
dfs_all = []
for timing_type in timing_types:
    dfs_timing_type = []
    for j in range(no_iterations):
        path = '%s/%s_%d.csv' %(logs_path, timing_type, j)
        df = pd.read_csv(path)
        df.rename({"Time":"Time_%s_%d" % (timing_type, j)}, axis=1, inplace=True)
        dfs_timing_type.append(df)
    df_stacked = pd.concat(dfs_timing_type)
    dfs_all.append(df_stacked)
    
dfs_all_stacked = []
for timing_type in timing_types:
    path_list = [i for i in glob.glob(('%s/%s*.{}' % (logs_path, timing_type)).format(extension))]
    df_stacked = pd.concat([pd.read_csv(f) for f in path_list])
    df_stacked.rename({"Time":"Time_%s" % timing_type}, axis=1, inplace=True)
    dfs_all_stacked.append(df_stacked)
    


df1 = pd.read_csv("../results/benchmark/default_baseline.csv")
df1.rename({"Time": "Baseline"}, axis=1, inplace=True)


fig = px.scatter(df1, x="Image count", y="Baseline", title="Processing time", trendline="rolling", 
                 trendline_options=dict(window=5))
fig.show()


for df in dfs_all :
    fig = px.scatter(df, x="Image count", y=df.columns[2:5], title="Processing time", trendline="rolling", 
                 trendline_options=dict(window=5))
    fig.show()
    
averaged_dfs = []
for df in dfs_all_stacked:
    averaged_dfs.append(df.groupby(["Image count"]).mean().reset_index())

all_stacked = pd.concat(averaged_dfs)

fig = px.scatter(all_stacked, x="Image count", y=all_stacked.columns[2:6], title="Processing time", trendline="rolling", 
                 trendline_options=dict(window=100))
fig.show()

# Total time
print("Baseline time: %s" % str(datetime.timedelta(seconds=df1["Baseline"].sum())))


```

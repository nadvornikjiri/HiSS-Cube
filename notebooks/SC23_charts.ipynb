```python
import pandas as pd
import matplotlib.pyplot as plt

path = "~/SDSSCube/results/benchmark/karolina_timings/all_phases_times.csv"

df  = pd.read_csv(path)

ax = pd.crosstab(index=df["CPUs"], columns=df["Phase"], values=df['Time']/df['Processed items'], normalize='index', aggfunc='sum').plot(title="Phase Time ratio - overall", kind='bar', stacked=True, rot=0)
lgd = ax.legend(title="Phase", bbox_to_anchor=(1, 1.02), loc='upper left')
ax.set_ylabel("Time [s]")
ax.set_xlabel("CPU cores")
plt.savefig('phase_ratio_overall.pdf', bbox_extra_artists=(lgd,), bbox_inches='tight')

df_without_ml_cube = df[df["Phase"] != "8 Global DB query"]

ax = pd.crosstab(index=df_without_ml_cube["CPUs"], columns=df_without_ml_cube["Phase"], values=df_without_ml_cube['Time']/df_without_ml_cube['Processed items'], normalize='index', aggfunc='sum').plot(
    title="Phase Time ratio - DB construction", kind='bar', stacked=True, rot=0)
lgd = ax.legend(title="Phase", bbox_to_anchor=(1, 1.02), loc='upper left')
ax.set_ylabel("Time [s]")
ax.set_xlabel("CPU cores")
plt.savefig('phase_ratio_db_buildup.pdf', bbox_extra_artists=(lgd,), bbox_inches='tight')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

path = "~/SDSSCube/results/benchmark/karolina_timings/image_data.csv"

df  = pd.read_csv(path)
df = df[df["IOC per node"] == "MPIO"]
df["Ratio"] = df["Total IO"]/df["File per process"]
ax = df.plot.bar(x="Nodes", y=["Total IO", "File per process"])
ax2 = df.plot.line(x="Nodes", y=["Ratio"], secondary_y=True, ax=ax, color="green")

ax.legend(["HiSS-Cube write bandwidth", "File per process write bandwidth"])
ax.set_ylabel("Bandwidth [MB/s]")
ax2.legend(["HiSS-Cube / File per process Ratio"], bbox_to_anchor=(0, 0.85), loc='upper left')
ax2.set_ylabel("Ratio")

plt.savefig('image_data.pdf')

print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

path = "~/SDSSCube/results/benchmark/karolina_timings/darshan.csv"

df  = pd.read_csv(path)
df["Other"] = df["Total"] - df["Write time"] - df["Read time"]
del df["Total"]
df = df.melt(id_vars=["Nodes", "Phase", "CPUs"], var_name="Type", value_name="Time")
df = df.sort_values(["Type","Time"], ascending=[False, True])
print(df)

order = ["Read time", "Write time", "Other"]



df2 = df[df["Phase"] == "5 Image data"]
print(pd.crosstab(index=df2["CPUs"], columns=df2["Type"], values=df2['Time'], aggfunc='sum')[order])
ax = (pd.crosstab(index=df2["CPUs"], columns=df2["Type"], values=df2['Time'], aggfunc='sum')[order]).plot(title="Cumulative time spent inside the image data phase", kind='bar', stacked=True, rot=0)
#lgd = ax.legend(title="Type", bbox_to_anchor=(1, 1.02), loc='upper left')
ax.set_ylabel("Time [s]")
ax.set_xlabel("CPU cores")

plt.savefig('darshan_image_data.pdf', bbox_inches='tight')

df2 = df[df["Phase"] == "6 Spectra data"]
print(pd.crosstab(index=df2["CPUs"], columns=df2["Type"], values=df2['Time'], aggfunc='sum')[order])
ax = (pd.crosstab(index=df2["CPUs"], columns=df2["Type"], values=df2['Time'], aggfunc='sum')[order]).plot(title="Cumulative time spent inside the spectra data phase", kind='bar', stacked=True, rot=0)
#lgd = ax.legend(title="Type", bbox_to_anchor=(1, 1.02), loc='upper left')
ax.set_ylabel("Time [s]")
ax.set_xlabel("CPU cores")

plt.savefig('darshan_spectra_data.pdf', bbox_inches='tight')

df2 = df[df["Phase"] == "8 Global DB query"]
print(pd.crosstab(index=df2["CPUs"], columns=df2["Type"], values=df2['Time'], aggfunc='sum')[order])
ax = (pd.crosstab(index=df2["CPUs"], columns=df2["Type"], values=df2['Time'], aggfunc='sum')[order]).plot(title="Cumulative time spent inside the Global DB Query phase", kind='bar', stacked=True, rot=0)
#lgd = ax.legend(title="Type", bbox_to_anchor=(1, 1.02), loc='upper left')
ax.set_ylabel("Time [s]")
ax.set_xlabel("CPU cores")

plt.savefig('darshan_db_query.pdf', bbox_inches='tight')

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path = "~/SDSSCube/results/benchmark/karolina_timings/darshan.csv"

df  = pd.read_csv(path)
df = df.melt(id_vars=["Nodes", "Phase", "CPUs"], var_name="Type", value_name="Time")
df = df.sort_values(["Time", "Phase", "CPUs", "Type"], ascending=[False, True, True, True])
print(df)

sns.relplot(kind='scatter', data=df, x='CPUs', y='Time', hue=df[['Phase', 'Type']].apply(tuple, axis=1))

plt.savefig('darshan_write.pdf', bbox_inches='tight')
```

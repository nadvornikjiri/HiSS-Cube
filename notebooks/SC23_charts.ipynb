```python
import pandas as pd
import matplotlib.pyplot as plt

path = "~/SDSSCube/results/benchmark/karolina_timings/all_phases_times.csv"

df  = pd.read_csv(path)

ax = pd.crosstab(index=df["CPUs"], columns=df["Phase"], values=df['Time']/df['Processed items'], normalize='index', aggfunc='sum').plot(title="Phase ratio - overall", kind='bar', stacked=True, rot=0)
lgd = ax.legend(title="Phase", bbox_to_anchor=(1, 1.02), loc='upper left')

plt.savefig('phase_ratio_overall.pdf', bbox_extra_artists=(lgd,), bbox_inches='tight')

df_without_ml_cube = df[df["Phase"] != "8 ML-cube"]

ax = pd.crosstab(index=df_without_ml_cube["CPUs"], columns=df_without_ml_cube["Phase"], values=df_without_ml_cube['Time']/df_without_ml_cube['Processed items'], normalize='index', aggfunc='sum').plot(
    title="Phase ratio - DB buildup", kind='bar', stacked=True, rot=0)
lgd = ax.legend(title="Phase", bbox_to_anchor=(1, 1.02), loc='upper left')

plt.savefig('phase_ratio_db_buildup.pdf', bbox_extra_artists=(lgd,), bbox_inches='tight')
```
